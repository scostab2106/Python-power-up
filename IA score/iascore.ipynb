```python
import pandas as pd
```

```python
#IMPORTING DATABASE
tabel = pd.read_csv('clientes.csv')
```

```python
#PREPAIR DATABASE FOR IA

    #TRANSFORM STR IN NUMBERS
        # - Label Encoder
from sklearn.preprocessing import LabelEncoder

codfier = LabelEncoder()
codfier2 = LabelEncoder()
codfier3 = LabelEncoder()
codfier4 = LabelEncoder()

tabel['profissao'] = codfier.fit_transform(tabel['profissao'])

tabel['mix_credito'] = codfier2.fit_transform(tabel['mix_credito'])

tabel['comportamento_pagamento'] = codfier3.fit_transform(tabel['comportamento_pagamento'])

tabel['score_credito'] = codfier4.fit_transform(tabel['score_credito'])


display(tabel.info())
```

```python

#DIVIDIR A BASE DE DADOS EM DOIS Y(QUAL VAI SER QUEM QUERO PREVER) E X(OS CARAS QUE VAO SER USADOS PARA PREVER O Y)

y = tabel['score_credito']
x = tabel.drop(columns=['score_credito'])

#DIVIDIR EM DADOS DE TREINO E DE TESTE

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y) #a ordem define pro sklearn saber qual é treino e qual é teste



```

```python
#CRIAR UM MODELO DE IA

#importando IA
from sklearn.ensemble import RandomForestClassifier #arvore de decisao
from sklearn.neighbors import KNeighborsClassifier  #vizinhos próximos

#cria IA
three_decision = RandomForestClassifier()
knn_model = KNeighborsClassifier()

#treina IA
three_decision.fit(x_train, y_train)
knn_model.fit(x_train, y_train)
```

```python
#AVALIAR QUAL É O MELHOR MODELO

previsao_three = three_decision.predict(x_test) #armazenando as previsoes da arvore na variavel
previsao_knn = knn_model.predict(x_test) #armazenando as previsoes da knn na variavel

#ACURÁCIO(PERCENTUAL CERTOS DAS PREVISOES)
from sklearn.metrics import accuracy_score #importando mecanismo de acuracia

display(accuracy_score(y_test, previsao_three))
display(accuracy_score(y_test, previsao_knn))
```
